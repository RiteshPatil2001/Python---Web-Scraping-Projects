#                                 WEB SCRPING ON A STOCK MARKET SITE

### ABSTRACT
#### Web scraping is the process of using bots to extract content and data from a website. Unlike screen scraping, which only copies pixels displayed onscreen, web scraping extracts the HTML code behind it. and, with it, data stored in a database. The scraper can then recreate entire website content elsewhere.

###### 

### LIBRARIES AND FUNCTIONS USED
#### In this web scraping project we are using 3 libraries which contain several function methods which we use throughout the project.

#### 1. bs4 - BeautifulSoup - BeautifulSoup is a Python library for pulling data out of HTML and XML files.
    From BesutifulSoup library we use 3 functions which are -
       i. prettify() - prettify function remove unnecessary content from scraping & returns the only HTML format of content.
      ii. find() - find function find the single searched entity.
     iii. find_all() - If we have number of entities with the same identification then we use this function.
      
#### 2. requests - The requests library is the de facto standard for making HTTP requests in Python.
    From request library we use 1 function which is -
       get() - get function is used to retrieve information from the given server using a given URL
      
#### 3. pandas - Pandas library is provides ready to use high-performance data structures and data analysis tools
    From pandas library we use 3 functions which are -
       i. DataFrame() - DataFrame function convert the all extrated data into table format.
      ii. to_csv() - This function is used to convert the tabular formated data into csv file which can we use in excel
     iii. read_csv() - This function is used to read the csv file in python.

###### 

### steps -
   ####  We need to import the libraries which we need 

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

#### Defing the URL

```python
url = "https://www.tickertape.in/screener/equity"
```

#### get the scraping response and content of url

```python
status = requests.get(url)
print(status)
```

```python
a = status.content
print(a)
```

#### 
#### prettify the content (remove unnecessary content)

```python
soup = BeautifulSoup(a, 'html.parser')
print(soup.prettify())
```

#### 
#### selecting the rquired field on the basis of their html class or id

```python
table4 = soup.find(id="screener-table")
table4
```

```python
sections = table4.find_all('section')
sections
```

```python
sec2 = sections[1]
sec2
```

#### 
#### fetch the data column wise from the section

### name column

```python
col1 = sec2.find(class_="jsx-1054294817 typography-body-medium-m stock-name-col fixed-col data-col text-left ellipsis-container d-flex-col")
print(col1)
```

```python
span_tag = col1('span')
span_tag
```

```python
company_tag = span_tag[2::3]
company_tag
```

```python
company_name = [i.get_text() for i in company_tag]
company_name
```

### sector column

```python
col2 = sec2.find(class_="jsx-1054294817 scroll-col data-col d-flex-col wider text-left ellipsis-container subindustry-col")
col2
```

```python
span_tag2 = col2('span')
span_tag2
```

```python
sector_tag = span_tag2[0::2]
sector_tag
```

```python
sector_name = [i.get_text() for i in sector_tag]
sector_name
```

### market cap column

```python
col3 = sec2.find(class_="jsx-1054294817 scroll-col data-col d-flex-col text-right mrktCapf-col")
col3
```

```python
span_tag3 = col3('span')
span_tag3
```

```python
market_cap_tag = span_tag3[0::2]
market_cap_tag
```

```python
market_cap = [i.get_text() for i in market_cap_tag]
market_cap
```

### Current Price column

```python
col4 = sec2.find(class_="jsx-1054294817 scroll-col data-col d-flex-col text-right lastPrice-col")
col4
```

```python
span_tag4 = col4('span')
span_tag4
```

```python
price_tag = span_tag4[0::2]
price_tag
```

```python
price = [i.get_text() for i in price_tag]
price
```

### PE return column

```python
col5 = sec2.find(class_="jsx-1054294817 scroll-col data-col d-flex-col text-right apef-col last-col")
col5
```

```python
span_tag5 = col5('span')
span_tag5
```

```python
pe_tag = span_tag5[0::2]
pe_tag
```

```python
pe_ratio = [i.get_text() for i in pe_tag]
pe_ratio
```

### 
### gather together all fetched data and store it into dictionary format

```python
data = {'Stock Name':company_name,'Sector':sector_name,'Market Cap (Cr)':market_cap,'Current Stock Price':price,'PE Ratio':pe_ratio}
data
```

### 
### Convert the data into table format

```python
table = pd.DataFrame(data)
table
```

### 
### Convert data into csv file (Excel File)

```python
table.to_csv('Stock Analysis.csv',index=False)
```

### Read the csv file

```python
read_file = pd.read_csv('Stock Analysis.csv')
read_file
```
